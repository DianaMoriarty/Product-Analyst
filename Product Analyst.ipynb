Вычисление среднего, максимального/минимального значений, медианы для всех числовых признаков для всего массива в целом.

# Самый популярный объект

datanew - новый датасет, в который включены объекты, у которых такие признаки как лайки, комментарии, подписчики и др. выше среднего. А так же только те пользователи, у которых paid = 1, поскольку именно у популярных пользователей чаще всего заказывают рекламу.

```python
f = ((data['Page total likes'] > data['Page total likes'].mean()) & 
     (data['Lifetime Post Total Reach'] > data['Lifetime Post Total Reach'].mean()) & 
     (data['Lifetime Engaged Users'] > data['Lifetime Engaged Users'].mean()) &
     (data['Lifetime Post Total Impressions'] > data['Lifetime Post Total Impressions'].mean()) &
     (data['Lifetime Post Consumers'] > data['Lifetime Post Consumers'].mean()) &
     (data['Lifetime Post Consumptions'] > data['Lifetime Post Consumptions'].mean()) &
     (data['comment'] > data['comment'].mean()) &
     (data['like'] > data['like'].mean()) &
     (data['share'] > data['share'].mean()) &
     (data['Total Interactions'] > data['Total Interactions'].mean()) &
     (data['Lifetime Post reach by people who like your Page'] > data['Lifetime Post reach by people who like your Page'].mean()) &
     (data['Lifetime Post Impressions by people who have liked your Page'] > data['Lifetime Post Impressions by people who have liked your Page'].mean()) &
     (data['Lifetime People who have liked your Page and engaged with your post'] > data['Lifetime People who have liked your Page and engaged with your post'].mean())) 
datanew = data.where(f).dropna()
datanew = datanew
datanew = datanew.where(datanew['Paid'] == 1.0).dropna()
datanew
```

Из получившихся девяти объектов сильно выделяется номер 244, особенно по признакам комментарии, репосты, общее количество людей, лайкнушвих страницу, из чего можно сделать вывод, что именно этот объект является самым популярным.
